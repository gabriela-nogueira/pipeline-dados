```python
import csv
```

```python
dados_csv = []

with open(path_csv, 'r') as file:
    spamreader = csv.reader(file)
    for row in spamreader:
        dados_csv.append(row)
```

```python
# Inicializa uma lista para armazenar os dados
dados = []

# Abre o arquivo CSV em modo de leitura
with open(path_csv, newline='') as arquivo_csv:
    # Cria um objeto leitor para o arquivo CSV
    leitor_csv = arquivo_csv.readlines()

    # Extrai os nomes das colunas (a primeira linha)
    colunas = leitor_csv[0].strip().split(',')

    # Itera sobre as linhas restantes e converte cada linha em um dicionário
    for linha in leitor_csv[1:]:
        valores = linha.strip().split(',')
        dicionario = dict(zip(colunas, valores))
        dados.append(dicionario)

# Exibe os dados lidos do arquivo CSV
print(dados)
```
